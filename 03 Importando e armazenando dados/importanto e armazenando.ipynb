```python
import os
```

```python
caminho_pasta = "dados/Sales/"

# cria uma lista com os arquivos da pasta 
arquivo_lista = os.listdir(caminho_pasta)
dataframes =[]


#ler cada arquivo csv dentro da pasta 
for arquivo in arquivo_lista:
    if arquivo.endswith(".csv"):

        #criando caminho completo para cada arquivo 
        arquivo_completo = os.path.join(caminho_pasta, arquivo)

        #importando cada arquivo csv para o df 
        df = pd.read_csv(arquivo_completo)

        #add o df na lista dataframe
        dataframes.append(df)

#contatenando todos dos dataframes dentro da lista em um unico dataframe
df_contatena = pd.concat(dataframes, ignore_index=True)
df_contatena
```
